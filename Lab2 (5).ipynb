### 3.3 Добавьте в таблицу `recipes` столбец `name_word_count`, в котором хранится количество слов из названии рецепта (считайте, что слова в названии разделяются только пробелами). Обратите внимание, что между словами может располагаться несколько пробелов подряд.

```python
import re
recipes["name_word_count"] = [len(re.sub("\s\s+", " ", i).split(" ")) for i in recipes["name"]]
recipes
```

### 5.1 При помощи объединения таблиц, создайте `DataFrame`, состоящий из четырех столбцов: `id`, `name`, `user_id`, `rating`. Рецепты, на которые не оставлен ни один отзыв, должны отсутствовать в полученной таблице. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и попытавшись найти строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
merged51 = recipes[["id", "name"]].merge(reviews[["user_id", "rating", "recipe_id"]], left_on="id", right_on="recipe_id", how="inner")
merged51 = merged51.drop(columns=["recipe_id"])
```

```python
merged51[merged51["id"]==342620] #проверка, рецепта без отзывов нет
```

### 5.2 При помощи объединения таблиц и группировок, создайте `DataFrame`, состоящий из трех столбцов: `recipe_id`, `name`, `review_count`, где столбец `review_count` содержит кол-во отзывов, оставленных на рецепт `recipe_id`. У рецептов, на которые не оставлен ни один отзыв, в столбце `review_count` должен быть указан 0. ### 5.3. Выясните, рецепты, добавленные в каком году, имеют наименьший средний рейтинг?

```python
merged53 = recipes[["id", "submitted", "name"]].merge(reviews.groupby("recipe_id")["rating"].mean().rename("rating-mean"), left_on="id", right_on="recipe_id", how="inner")
merged53["submitted"]=pd.DatetimeIndex(merged53["submitted"]).year
merged53 = merged53.sort_values(by="rating-mean")
merged53
```
